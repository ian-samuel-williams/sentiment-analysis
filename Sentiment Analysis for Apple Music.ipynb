```python
songNames = ["Fast Lane", "Welcome to the Party", "Hello", "Lemonade", "Soul Food"]
songSentiments = pd.DataFrame(index = songNames, columns = ["Positive", "Neutral", "Negative"])
songSentiments
```

```python
songs = [Fast_Lane, Welcome_to_the_Party, Hello, Lemonade, Soul_Food]
```

```python
analyzer = SentimentIntensityAnalyzer()
songPositives = []
songNeutrals = []
songNegatives = []
for song in songs:
    positive = 0
    negative = 0
    neutral = 0
    for i in range(0, len(song)):
        ps = analyzer.polarity_scores(song[i])
        if ps["compound"] >= 0.05:
            positive += 1
        elif ps["compound"] <= -0.05:
            negative += 1
        else:
            neutral += 1
    songPositivesper = (positive / (i + 1)) * 100
    songNegativesper = (negative / (i + 1)) * 100
    songNeutralsper = (neutral / (i + 1)) * 100
    songPositives.append(songPositivesper)
    songNegatives.append(songNegativesper)
    songNeutrals.append(songNeutralsper)
```

```python
songSentiments["Positive"] = songPositives
songSentiments["Neutral"] = songNeutrals
songSentiments["Negative"] = songNegatives
songSentiments
```
